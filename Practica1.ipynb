##Preparació de dades

```python
import pandas as pd
import numpy as np
```

```python
# LO SUBI A GITHUB PARA QUE PODAMOS ACCEDER LOS 2
df=pd.read_csv('https://raw.githubusercontent.com/GLuarte/IAAI/main/AIRE_BARCELONA_2018.csv',sep=';')
df
```

```python
#Es irrelevante tener el "Barcelona" en el nombre de la cabina pues todos los datos son de Barcelona, por lo que se elimina
df.nom_cabina=[item.replace('Barcelona - ','') for item in df.nom_cabina]
```

```python
df.iloc[405:410, :]
```

```python
# '--' a NaN
df["qualitat_aire"].replace({'--':np.NaN}, regex=True, inplace=True)

#Se quitan las unidades de los datos de las celdas pues no son necesarios para el análisis y los '--' se pasan a NaN por facilidad en el manejo de la data, pues '--' no genera información relevante
#Para o3
df['hora_o3'].replace({'h':''}, regex=True, inplace=True) #Quitar la unidad de las horas
df['qualitat_o3'].replace({'--':np.NaN}, regex=True, inplace=True) # Pasar '--' a NaN
df["valor_o3"].replace({'µg/m³':'', ' ': '', '--':np.NaN}, regex=True, inplace=True) # Quitar unidades, quitar espacios indeseados, '--' a NaN
df["valor_o3"].replace({'': '0', '--':np.NaN}, regex=True, inplace=True) # No sé si dejar esto
#Para no2
df['hora_no2'].replace({'h':''}, regex=True, inplace=True) 
df['qualitat_no2'].replace({'--':np.NaN}, regex=True, inplace=True)
df["valor_no2"].replace({'µg/m³':'', ' ': '', '--':np.NaN}, regex=True, inplace=True)
#Para pm10
df['hora_pm10'].replace({'h':''}, regex=True, inplace=True)
df['qualitat_pm10'].replace({'--':np.NaN}, regex=True, inplace=True)
df["valor_pm10"].replace({'µg/m³':'', ' ': '', '--':np.NaN}, regex=True, inplace=True)

#df.valor_no2=[item.replace(' µg/m³','') for item in df.valor_no2]

#np.where(df['valor_no2 (µg/m³)'].isnull())

#Falta convertir a int!!!!!
```

```python
# '--' a NaN
df["qualitat_aire"].replace({'--':np.NaN}, regex=True, inplace=True)

#Se quitan las unidades de los datos de las celdas pues no son necesarios para el análisis y los '--' se pasan a NaN por facilidad en el manejo de la data, pues '--' no genera información relevante
#Para o3
l=["o3","no2","pm10"]
for i in l:
  df['hora_'+i].replace({'h':''}, regex=True, inplace=True) #Quitar la unidad de las horas
  df['qualitat_'+i].replace({'--':np.NaN}, regex=True, inplace=True) # Pasar '--' a NaN
  df["valor_"+i].replace({'µg/m³':'', ' ': '', '--':np.NaN}, regex=True, inplace=True) # Quitar unidades, quitar espacios indeseados, '--' a NaN
  df["valor_"+i].replace({'': '0', '--':np.NaN}, regex=True, inplace=True) # No sé si dejar esto
  df=df.astype({"valor_"+i: 'float'})
  
  #df.valor_no2=[item.replace(' µg/m³','') for item in df.valor_no2]

#np.where(df['valor_no2 (µg/m³)'].isnull())

#Falta convertir a int!!!!!
#Converti a float pq int no soporta los NaN
```

```python
df=df.astype({'valor_o3': 'float'})
```
